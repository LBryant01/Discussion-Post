```python
#Webscrape
import requests
response = requests.get('https://www.python.org')
response.content
```

```python
#Extract only text from the page
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content,'html5lib')
text = soup.get_text(strip=True)
```

```python
#Eliminate the stop words
from pathlib import Path
from nltk.corpus import stopwords
from textblob import TextBlob
blob=TextBlob(text)
stop_words = stopwords.words('english')
items = blob.word_counts.items()
items = [item for item in items if item[0] not in stop_words]
```

```python
#Use word cloud module which removes stop words from the text argument, using the wordcloud module’s built-in stop-words list
from wordcloud import WordCloud
import imageio
%matplotlib inline
image_file="https://media.cheggcdn.com/media/216/21621ee5-e80f-47f3-9145-513f2229b390/phploeBuh.png"
mask_image = imageio.imread(image_file)
wordcloud = WordCloud(width=1000, height=1000,
    colormap='prism', mask=mask_image, background_color='white')

wordcloud = wordcloud.generate(text)
wordcloud = wordcloud.to_file('text.png')
import matplotlib.pyplot as plt
plt.axis('off')
plt.imshow(wordcloud)
```
